```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
height=np.random.normal(1.75,0.20,5000)
weight=np.random.normal(60.32,15,5000)
print(height)
print(weight)
```

```python
np_height=np.array(height)
np_weight=np.array(weight)
np_inclued=np.array([np_height,np_weight])
```

```python
print("Mean"+" "+ str(np.mean(np_inclued[0])))
```

```python
print("Median"+" "+ str(np.median(np_inclued[1])))
```

```python
print("Correfficent"+ " "+str(np.corrcoef(np_inclued[0],np_inclued[1])))
```

```python
plt.plot(height,weight)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.show()
```

```python
plt.scatter(height,weight)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.grid(True)
plt.show()
```

```python
plt.hist(height,bins=10)
plt.show()
```
